```python
import requests
import pandas as pd
import io
import datetime

# API endpoint and headers
url = "https://visual-crossing-weather.p.rapidapi.com/history"
headers = {
    "X-RapidAPI-Key": "8313dc4870msh8ab9152b95b5effp19177ajsn329593311cc2",
    "X-RapidAPI-Host": "visual-crossing-weather.p.rapidapi.com"
}

# Set start and end date
start_date = datetime.date(2019, 1, 1)
end_date = datetime.date.today()

# Set location and query parameters
location = "Warangal"
params = {
    "aggregateHours": "24",
    "unitGroup": "us",
    "location": location,
    "shortColumnNames": "True",
    "contentType": "csv"
}

# Initialize an empty DataFrame to hold the data
all_data = pd.DataFrame()

# Loop through dates and fetch data
date = start_date
while date <= end_date:
    # Set query parameters for current date
    params["startDateTime"] = date.strftime("%Y-%m-%dT%H:%M:%S")
    params["endDateTime"] = (date + datetime.timedelta(days=30)).strftime("%Y-%m-%dT%H:%M:%S")

    # Make API request and get response as string
    response = requests.get(url, headers=headers, params=params)
    response.raise_for_status()
    response_str = response.text

    # Convert response string to pandas DataFrame
    df = pd.read_csv(io.StringIO(response_str))

    # Append current month's data to the all_data DataFrame
    all_data = all_data.append(df, ignore_index=True)

    # Move to next month
    date = date + datetime.timedelta(days=30)

# Save DataFrame to CSV file
filename = f"{location}_{start_date.strftime('%Y-%m-%d')}_{end_date.strftime('%Y-%m-%d')}.csv"
all_data.to_csv(filename, index=False)
```
